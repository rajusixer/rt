## Concept
Web Scraping is the automated process of extracting data from websites.
- **HTML Scraping**: Parses HTML pages using BeautifulSoup to pull tables, text, and tags.
- **JSON Scraping**: Reads structured JSON data returned by REST APIs using Python's `json` module.
Both approaches are essential for collecting real-world data when no ready-made dataset exists.

## Libraries & Functions Used
| Library/Function | Purpose |
|---|---|
| `pandas` | Create & manipulate DataFrames |
| `beautifulsoup4` | Parse HTML and locate tags |
| `requests` | Send HTTP GET requests to fetch web pages |
| `urllib.request.urlopen` | Open a URL and read raw response |
| `json.loads()` | Convert JSON string → Python list/dict |
| `soup.find('table')` | Locate the first HTML table |
| `row.find_all('td')` | Get all data cells in a table row |
| `pd.DataFrame()` | Build a structured table from lists |
| `df.head(10)` | Preview first 10 rows |

## Code Flow
```
1. Install pandas & beautifulsoup4
2. Send HTTP request with browser-like headers
3. Parse response HTML with BeautifulSoup
4. Find the table → loop through rows → extract cells
5. Store Sr.No, Country, Area in lists → build DataFrame
── JSON Part ──
6. Open API URL with urlopen
7. json.loads() → Python list of dicts
8. Loop through dicts → extract Id, Username, Email
9. Build DataFrame → display
```

## Part A – HTML Scraping

````python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
           'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

response = requests.get(url, headers=headers)
page = response.text
soup = BeautifulSoup(page, 'html.parser')
table = soup.find('table')
print("Table found:", table is not None)
````

````python
SrNo, Country, Area = [], [], []

rows = table.find('tbody').find_all('tr')
for row in rows:
    cells = row.find_all('td')
    if cells:
        SrNo.append(cells[0].get_text().strip('\n'))
        Country.append(cells[1].get_text().strip('\xa0').strip('\n'))
        Area.append(cells[3].get_text().strip('\n'))

print("Extracted rows:", len(SrNo))
````

````python
df = pd.DataFrame()
df['SrNo']    = SrNo
df['Country'] = Country
df['Area']    = Area
df.head(10)
````

## Part B – JSON Scraping

````python
from urllib.request import urlopen
import json

url_json = 'https://jsonplaceholder.typicode.com/users'
page = urlopen(url_json)
data = json.loads(page.read())

Id, Username, Email = [], [], []
for item in data:
    Id.append(item.get('id', 'NA'))
    Username.append(item.get('username', 'NA'))
    Email.append(item.get('email', 'NA'))

df_json = pd.DataFrame({'Id': Id, 'Username': Username, 'Email': Email})
df_json.head(10)
````
