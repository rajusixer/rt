```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
boston_df = pd.read_csv("Boston.csv")
boston_df.info()
boston_df.head()
```

```python
boston_df = boston_df.drop("Unnamed: 0", axis=1)
boston_df.info()
```

```python
boston_x = pd.DataFrame(boston_df.iloc[:, :13])
boston_y = pd.DataFrame(boston_df.iloc[:, -1])

X_train, X_test, Y_train, Y_test = train_test_split(boston_x, boston_y, test_size=0.3)
print(f"X Train size: {X_train.shape}")
print(f"X Test size: {X_test.shape}")
print(f"Y Train size: {Y_train.shape}")
print(f"Y Test size: {Y_test.shape}")
```

```python
reg_model = LinearRegression()
reg_model.fit(X_train, Y_train)
y_predicted = reg_model.predict(X_test)

Y_pred = pd.DataFrame(y_predicted, columns=["Predicted_Y"])
Y_pred.head()
```

```python
plt.scatter(Y_test, Y_pred, c="black")
plt.xlabel("Actual Price (medv)")
plt.ylabel("Predicted Price")
plt.show()
```

This notebook reads the Boston housing dataset, removes the unwanted index column, and splits the data into input features and target output.  
A multivariate linear regression model is then trained using many housing features together.  
The final scatter plot compares actual house prices with predicted prices to show how well the model performs.
