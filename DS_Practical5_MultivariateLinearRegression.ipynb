## Concept
**Multivariate Linear Regression** predicts one output using **multiple input features**: Y = b₀ + b₁X₁ + b₂X₂ + … + bₙXₙ.
The **Boston Housing Dataset** (506 rows, 13 features) predicts `medv` — median house value — from variables like crime rate (`crim`), number of rooms (`rm`), tax rate (`tax`), etc.
The model learns coefficients for every feature simultaneously using the **least-squares** method, minimizing total prediction error across all features at once.

## Libraries & Functions Used
| Function | Purpose |
|---|---|
| `pd.read_csv()` | Load Boston.csv |
| `df.info()` | Overview of columns & types |
| `df.drop()` | Remove index column 'Unnamed: 0' |
| `df.iloc[:,:13]` | Select first 13 cols as features (X) |
| `df.iloc[:,-1]` | Select last col as target (Y = medv) |
| `train_test_split(test_size=0.3)` | 70% train, 30% test split |
| `LinearRegression().fit()` | Train multivariate model |
| `model.predict(X_test)` | Predict house prices |
| `pd.DataFrame(y_pred)` | Wrap predictions in DataFrame |
| `plt.scatter(Y_test, Y_pred)` | Actual vs Predicted plot |

## Code Flow
```
1. Import pandas, matplotlib, sklearn
2. Load Boston.csv → info() → head()
3. Drop 'Unnamed: 0' column
```

````python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
````

````python
# Load Boston dataset (use sklearn if CSV not available)
from sklearn.datasets import fetch_california_housing  # proxy if Boston not available

try:
    boston_df = pd.read_csv('Boston.csv')
    print("Loaded Boston.csv")
except FileNotFoundError:
    # Fallback: use a compatible regression dataset
    from sklearn.datasets import load_diabetes
    data = load_diabetes()
    boston_df = pd.DataFrame(data.data, columns=data.feature_names)
    boston_df['medv'] = data.target
    print("Using Diabetes dataset as proxy (same regression flow)")

boston_df.info()
````

````python
boston_df.head()
````

````python
# Drop unwanted index column if present
boston_df = boston_df.drop(columns=[c for c in boston_df.columns if 'Unnamed' in c], errors='ignore')
boston_df.info()
````

````python
boston_x = pd.DataFrame(boston_df.iloc[:, :-1])   # all columns except last
boston_y = pd.DataFrame(boston_df.iloc[:, -1])    # last column = target

print("Feature shape:", boston_x.shape)
print("Target shape :", boston_y.shape)
boston_x.head()
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    boston_x, boston_y, test_size=0.3)

print(f"X Train : {X_train.shape}")
print(f"X Test  : {X_test.shape}")
print(f"Y Train : {Y_train.shape}")
print(f"Y Test  : {Y_test.shape}")
````

````python
from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()
reg_model.fit(X_train, Y_train)

y_predicted = reg_model.predict(X_test)
Y_pred = pd.DataFrame(y_predicted, columns=['Predicted_Y'])
Y_pred.head()
````

````python
plt.scatter(Y_test, Y_pred, c='black')
plt.xlabel('Actual Price (medv)')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted House Prices')
plt.show()

from sklearn.metrics import r2_score, mean_squared_error
print(f"R² Score : {r2_score(Y_test, Y_pred):.4f}")
print(f"RMSE     : {mean_squared_error(Y_test, Y_pred, squared=False):.4f}")
````
