## Concept
**Univariate (Simple) Linear Regression** is a supervised ML algorithm that predicts a continuous output from a single input feature by fitting the best-fit line: **Y = b₀ + b₁X + ε**.
- **4A**: Predict employee **Salary** from **Years of Experience** using a synthetic dataset.
- **4B**: Simulate the model **Y = 10 + 7X + ε** on 100 random samples and visualize predictions.
Scikit-learn's `LinearRegression` learns b₀ (intercept) and b₁ (slope) by minimizing the sum of squared residuals.

## Libraries & Functions Used
| Function | Purpose |
|---|---|
| `datasets.make_regression()` | Generate synthetic regression data |
| `np.interp()` | Rescale array to a new numeric range |
| `LinearRegression()` | Create a regression model object |
| `model.fit(X, y)` | Train model — learn slope & intercept |
| `model.predict(X)` | Generate predicted Y values |
| `np.random.rand(100,1)` | Create 100 random samples |
| `plt.plot()` | Draw scatter or line plots |
| `plt.legend()` | Display plot legend |

## Code Flow

## Part 4A – Salary Prediction

````python
import numpy as np
!pip install scikit-learn
````

````python
from sklearn import datasets
import matplotlib.pyplot as plt

x, y, coef = datasets.make_regression(
    n_samples=100, n_features=1, n_informative=1,
    noise=10, coef=True, random_state=0)

# Scale to realistic ranges
x = np.interp(x, (x.min(), x.max()), (0, 20))   # 0–20 years experience
y = np.interp(y, (y.min(), y.max()), (20000, 150000))  # salary in ₹

plt.plot(x, y, '.', label='Training Data')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary – Raw Data')
plt.legend()
plt.show()
````

````python
from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()
reg_model.fit(x, y)
y_predicted = reg_model.predict(x)

plt.plot(x, y, '.', label='Training Data', color='orange')
plt.plot(x, y_predicted, '.', color='black', label='Predicted')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary – With Regression Line')
plt.legend()
plt.show()

print(f"Slope (b1)     : {reg_model.coef_[0]:.2f}")
print(f"Intercept (b0) : {reg_model.intercept_:.2f}")
````

## Part 4B – Simulate Y = 10 + 7X + ε

````python
from sklearn.linear_model import LinearRegression
import numpy as np, matplotlib.pyplot as plt

reg_model2 = LinearRegression()
x2 = np.random.rand(100, 1)
error = np.random.rand(100, 1)
b0, b1 = 10, 7
y2 = b0 + b1 * x2 + error

reg_model2.fit(x2, y2)
y_pred2 = reg_model2.predict(x2)

plt.plot(x2, y_pred2, '.', color='black', label='Predicted Data')
plt.plot(x2, y2, '.', label='Training Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simulated Y = 10 + 7X + ε')
plt.legend()
plt.show()

print(f"Learned b0 : {reg_model2.intercept_[0]:.4f}  (true: 10)")
print(f"Learned b1 : {reg_model2.coef_[0][0]:.4f}  (true: 7)")
````
