## Concept
The **MTCars dataset** (Motor Trend Car Road Tests) contains specs for 32 car models (1973–74).
It has 12 variables: mpg, cyl, hp, wt, am (transmission), gear, etc.
EDA on MTCars shows that **heavier cars with more cylinders get lower fuel efficiency (MPG)** and that **manual transmission cars are more fuel-efficient**.
This dataset is a classic for learning correlation analysis and multivariate visualization.

## Libraries & Functions Used
| Function | Purpose |
|---|---|
| `pd.read_csv()` | Load MTCars CSV |
| `df.head() / tail()` | Preview first/last rows |
| `df.info()` | Structure, dtypes, null counts |
| `df.describe()` | Summary statistics |
| `df.isnull().sum()` | Missing value check |
| `df.drop()` | Remove 'model' text column |
| `fillna(groupby…transform)` | Group-wise mean imputation for mpg |
| `sns.countplot()` | Cylinder distribution |
| `plt.hist()` | MPG frequency histogram |
| `sns.boxplot()` | HP spread + outlier detection |
| `sns.catplot()` | Cyl vs Transmission count |
| `sns.violinplot()` | MPG by cyl split by am |
| `sns.scatterplot()` | wt vs mpg coloured by cyl, sized by hp |
| `df.corr(method='pearson')` | Correlation matrix |
| `sns.heatmap(cmap='coolwarm')` | Coloured correlation heatmap |

````
import os, pandas as pd, seaborn as sns
import matplotlib.pyplot as plt, numpy as np

# Download MTCars if not present
url = 'https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv'
df = pd.read_csv(url)
print(df.head())
print(df.tail())
````

````
df.info()
print(df.isnull().sum())
print(df.describe())
print(df.shape)
````

````
df_cleaned = df.drop(['model'], axis=1, errors='ignore')
df_cleaned['mpg'] = df_cleaned['mpg'].fillna(
    df_cleaned.groupby('cyl')['mpg'].transform('mean'))
print("Missing values after cleaning:")
print(df_cleaned.isnull().sum())
````

````
plt.figure(figsize=(8,5))
sns.countplot(x='cyl', data=df_cleaned)
plt.title('Distribution of Cars by Cylinder Count')
plt.show()
````

````
plt.figure(figsize=(8,5))
plt.hist(df_cleaned['mpg'], bins=10, color='skyblue', edgecolor='black')
plt.title('MPG Frequency Histogram')
plt.xlabel('MPG'); plt.ylabel('Frequency')
plt.show()
````

````
plt.figure(figsize=(8,5))
sns.boxplot(x=df_cleaned['hp'])
plt.title('Horsepower Boxplot (Outlier Detection)')
plt.show()
````

````
sns.catplot(x='cyl', hue='am', kind='count', data=df_cleaned)
plt.title('Cylinder vs Transmission Type')
plt.show()
````

````
group1 = df_cleaned.groupby(['cyl','am'])
cyl_am_table = group1.size().unstack()
sns.heatmap(cyl_am_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Cyl × AM Heatmap')
plt.show()
````

````
plt.figure(figsize=(10,6))
sns.violinplot(x='cyl', y='mpg', hue='am', data=df_cleaned, split=True)
plt.title('MPG Distribution by Cylinders & Transmission')
plt.show()
````

````
plt.figure(figsize=(10,6))
sns.scatterplot(x='wt', y='mpg', hue='cyl', size='hp', data=df_cleaned)
plt.title('Weight vs MPG (hue=Cyl, size=HP)')
plt.show()
````

````
mtcars_corr = df_cleaned.corr(method='pearson')
plt.figure(figsize=(12,8))
sns.heatmap(mtcars_corr, annot=True, vmax=1, vmin=-1, cmap='coolwarm')
plt.title('Pearson Correlation Heatmap – MTCars')
plt.show()
print(mtcars_corr)
````
