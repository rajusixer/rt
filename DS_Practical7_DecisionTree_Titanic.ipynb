## Concept
A **Decision Tree** is a supervised classification algorithm that splits data into branches based on feature values, forming a tree-like structure of if-then rules.
Splitting quality is measured by **Entropy** (information gain): the model picks the feature and threshold that best separates the classes.
`max_depth=4` limits tree growth to prevent overfitting.
On the Titanic dataset the tree learns rules like: *"If Sex=female and Pclass=1 → Survived"*, making it highly interpretable.

## Libraries & Functions Used

````
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
````

````
# Load Titanic dataset (sklearn fetch as fallback)
try:
    titanic_df = pd.read_csv('train.csv')
    print("Loaded train.csv")
except FileNotFoundError:
    from sklearn.datasets import fetch_openml
    raw = fetch_openml('titanic', version=1, as_frame=True)
    titanic_df = raw.frame.copy()
    titanic_df.columns = [c.strip().capitalize() for c in titanic_df.columns]
    titanic_df.rename(columns={'Sibsp':'SibSp','Pclass':'Pclass',
                               'Sex':'Sex','Age':'Age','Parch':'Parch',
                               'Survived':'Survived'}, inplace=True)
    titanic_df['Survived'] = titanic_df['survived'].astype(int)
    print("Using OpenML Titanic dataset")

titanic_df.head()
````

````
titanic_df.info()
````

````
features = ['pclass','sex','age','sibsp','parch']
# normalise column names
titanic_df.columns = titanic_df.columns.str.lower()
X = titanic_df[features].copy()
print(X.info())
X.head()
````

````
Y = titanic_df['survived']
Y = Y.astype(int)
print(Y.value_counts())
````

````
# Encode Sex
X['sex'] = X['sex'].map({'male': 0, 'female': 1})
# Fill missing Age
X['age'] = X['age'].fillna(X['age'].median())
print("Nulls remaining:", X.isnull().sum().sum())
X.head()
````

````
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=10)

dtmodel = DecisionTreeClassifier(
    criterion='entropy', max_depth=4, random_state=10)
dtmodel.fit(X_train, Y_train)
````

````
Y_pred = dtmodel.predict(X_test)
print("Accuracy:", accuracy_score(Y_test, Y_pred))
print()
print(classification_report(Y_test, Y_pred,
      target_names=['Not Survived','Survived']))
````

````
plt.figure(figsize=(18, 10))
plot_tree(dtmodel,
          feature_names=features,
          class_names=['Not Survived', 'Survived'],
          filled=True, rounded=True)
plt.title('Decision Tree – Titanic Survival')
plt.show()
````
