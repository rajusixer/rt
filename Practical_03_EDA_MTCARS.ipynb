```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

print(os.getcwd())
df = pd.read_csv("MTCARS.csv")
print(df.head())
print(df.tail())
df.info()
print(df.isnull().sum())
print(df.describe())
print(df.shape)
```

```python
df_cleaned = df.drop(["model"], axis=1)
df_cleaned["mpg"] = df_cleaned["mpg"].fillna(df_cleaned.groupby("cyl")["mpg"].transform("mean"))
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="cyl", data=df_cleaned)
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df_cleaned["mpg"], bins=10)
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned["hp"])
plt.show()
```

```python
sns.catplot(x="cyl", hue="am", kind="count", data=df_cleaned)
plt.show()

group1 = df_cleaned.groupby(["cyl", "am"])
cyl_am_table = group1.size().unstack()
sns.heatmap(cyl_am_table, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="cyl", y="mpg", hue="am", data=df_cleaned, split=True)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x="wt", y="mpg", hue="cyl", size="hp", data=df_cleaned)
plt.show()
```

```python
mtcars_corr = df_cleaned.corr(method="pearson")
print(mtcars_corr)

plt.figure(figsize=(12, 8))
sns.heatmap(mtcars_corr, annot=True, vmax=1, vmin=-1, cmap="coolwarm")
plt.show()
```

This notebook reads the MTCARS dataset, checks its rows, columns, data types, summary statistics, and missing values.  
After removing the text `model` column and filling missing `mpg` values using cylinder-wise averages, it creates several visualizations.  
The final correlation heatmap helps show how features like weight, horsepower, and mileage are connected.
