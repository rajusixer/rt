```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
cleaned = df.drop(["PassengerId", "Name", "Ticket", "Fare", "Cabin"], axis=1)
cleaned.head()
```

```python
cleaned["Age"] = cleaned["Age"].fillna(cleaned.groupby("Pclass")["Age"].transform("mean"))
cleaned.isnull().sum()
```

```python
sns.catplot(x="Sex", hue="Survived", kind="count", data=cleaned)
plt.show()
```

```python
gender_survived = cleaned.groupby(["Sex", "Survived"]).size().unstack()
sns.heatmap(gender_survived, annot=True, fmt="d")
plt.show()
```

```python
pclass_survived = cleaned.groupby(["Pclass", "Survived"]).size().unstack()
sns.heatmap(pclass_survived, annot=True, fmt="d")
plt.show()
```

```python
sns.violinplot(x="Sex", y="Age", hue="Survived", data=cleaned, split=True)
plt.show()
```

```python
cleaned_corr = cleaned.drop(["Sex", "Embarked"], axis=1)
print(cleaned_corr.corr(method="pearson"))

sns.heatmap(cleaned_corr.corr(method="pearson"), annot=True, vmax=1)
plt.show()
```

This practical loads the Titanic dataset, checks its structure, and finds missing values.  
It removes less useful columns, fills missing age values using the average age within each passenger class, and then creates different plots.  
The charts help compare survival with sex, class, and age, while the correlation heatmap shows how numerical features are related.
