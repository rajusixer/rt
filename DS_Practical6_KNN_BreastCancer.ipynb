## Concept
**K-Nearest Neighbors (KNN)** is a supervised classification algorithm that assigns a class label to a new point by finding the **K closest training samples** (using Euclidean distance) and taking a majority vote.
The **Breast Cancer Wisconsin Dataset** (569 samples, 30 features) classifies tumors as **Benign (1)** or **Malignant (0)**.
We use only 2 features — `mean area` and `mean compactness` — for easy 2D visualization.
Model performance is evaluated via **Confusion Matrix, Accuracy, Precision, Recall, and F1-Score**.

## Libraries & Functions Used
| Function | Purpose |
|---|---|
| `load_breast_cancer()` | Built-in sklearn breast cancer dataset |
| `pd.DataFrame(data, columns=names)` | Convert numpy array → DataFrame |
| `X_df[['mean area','mean compactness']]` | Select 2 features |
| `pd.Categorical.from_codes()` | Map 0/1 → 'malignant'/'benign' labels |
| `pd.get_dummies(drop_first=True)` | One-hot encode → binary column |
| `train_test_split(random_state=1)` | Reproducible train/test split |
| `KNeighborsClassifier(n_neighbors=5, metric='euclidean')` | Build KNN model |
| `knn.fit()` | Train model |
| `knn.predict()` | Generate predictions |
| `confusion_matrix()` | TP, TN, FP, FN table |
| `sns.heatmap()` | Visualise confusion matrix |
| Manual formulas | Accuracy, Precision, Recall, F1 |

## Code Flow
```
1. Import numpy, pandas, matplotlib, seaborn, sklearn
2. load_breast_cancer() → X_df (all 30 features)
3. Select 2 features: mean area, mean compactness
4. Encode target: Categorical → one-hot (benign=True/False)
5. train_test_split(random_state=1)
6. KNeighborsClassifier(k=5, euclidean).fit(X_train, Y_train)
7. Scatter plot: actual class labels on test data
8. Y_pred = knn.predict(X_test)
9. Scatter plot: predicted class labels (coolwarm colours)
10. confusion_matrix(Y_test, Y_pred) → print + heatmap
11. Extract TP,FN,FP,TN → compute metrics
    Accuracy = (TP+TN)/(TP+TN+FP+FN)
    Precision = TP/(TP+FP)
    Recall    = TP/(TP+FN)
    F1        = 2*(P*R)/(P+R)
```

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
````

````
sns.set()
breast_cancer_data = load_breast_cancer()

X_df = pd.DataFrame(breast_cancer_data.data,
                    columns=breast_cancer_data.feature_names)
X_df.head()
````

````
# Select 2 features for 2-D visualization
X_df = X_df[['mean area', 'mean compactness']]
X_df.head()
````

````
Y_df = pd.Categorical.from_codes(
    breast_cancer_data.target, breast_cancer_data.target_names)
print(Y_df)
````

````
Y_df = pd.get_dummies(Y_df, drop_first=True)
print(Y_df)
````

````
X_train, X_test, Y_train, Y_test = train_test_split(
    X_df, Y_df, random_state=1)

knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train, Y_train)
````

````
# Scatter: actual labels
combined_df = pd.concat([X_test, Y_test], axis=1)
sns.scatterplot(x='mean area', y='mean compactness',
                hue='benign', data=combined_df)
plt.title('Actual Labels – Test Set')
plt.show()
````

````
# Scatter: predicted labels
Y_pred = knn.predict(X_test)
plt.scatter(X_test['mean area'], X_test['mean compactness'],
            c=Y_pred, cmap='coolwarm', alpha=0.7)
plt.xlabel('mean area'); plt.ylabel('mean compactness')
plt.title('KNN Predicted Labels (coolwarm)')
plt.show()
````

````
# Confusion Matrix
cf = confusion_matrix(Y_test, Y_pred)
print(cf)

ax = plt.subplot()
sns.heatmap(cf, annot=True, ax=ax)
ax.set_xlabel('Predicted Value')
ax.set_ylabel('Actual Value')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['Malignant', 'Benign'])
ax.yaxis.set_ticklabels(['Malignant', 'Benign'])
plt.show()
````

````
# Performance Metrics
tp, fn, fp, tn = confusion_matrix(Y_test, Y_pred, labels=[1,0]).reshape(-1)
print(f"TP={tp}  FN={fn}  FP={fp}  TN={tn}")

accuracy  = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall    = tp / (tp + fn)
f1_score  = 2 * (precision * recall) / (precision + recall)

print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1_score:.4f}")
````
