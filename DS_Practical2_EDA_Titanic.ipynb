## Concept
**Exploratory Data Analysis (EDA)** is the first step before applying any ML model.
It means "getting to know your data" — checking its shape, missing values, distributions and relationships.
The **Titanic dataset** (891 rows, 12 columns) records passenger details and survival outcomes from the 1912 disaster.
EDA on Titanic reveals that gender and passenger class are the strongest survival predictors.

## Libraries & Functions Used
| Function | Purpose |
|---|---|
| `pd.read_csv()` | Load CSV into DataFrame |
| `df.info()` | Column types, null counts, memory usage |
| `df.describe()` | Statistical summary (mean, std, min, max) |
| `df.isnull().sum()` | Count missing values per column |
| `df.drop()` | Remove unwanted columns |
| `fillna(groupby…mean)` | Impute Age with class-wise mean |
| `sns.catplot()` | Count-plot by category |
| `sns.heatmap()` | Correlation / cross-tab heatmap |
| `sns.violinplot()` | Age distribution split by survival & sex |

## Code Flow
```
1. Import numpy, pandas, matplotlib, seaborn
2. Load train.csv → df.head(), df.info()
3. df.describe() → df.isnull().sum()
4. Drop irrelevant cols (PassengerId, Name, Ticket, Fare, Cabin)
5. Fill missing Age with group-mean by Pclass
6. catplot → survival count by Sex
7. Heatmap → survival count by Sex × Survived
8. Heatmap → survival count by Pclass × Survived
9. Violin plot → Age distribution by Sex & Survived
10. Pearson correlation matrix → heatmap
```

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Load dataset (place train.csv in working directory)
import os
# df = pd.read_csv('train.csv')   # uncomment when file is present

# Demo: create a small synthetic Titanic-like DataFrame
from sklearn.datasets import fetch_openml
titanic_raw = fetch_openml('titanic', version=1, as_frame=True)
df = titanic_raw.frame.copy()
df.columns = [c.capitalize() for c in df.columns]
# Rename to match journal columns
rename = {'Pclass':'Pclass','Sex':'Sex','Age':'Age',
          'Sibsp':'SibSp','Parch':'Parch','Survived':'Survived'}
df.rename(columns={'Sibsp':'SibSp'}, inplace=True)
print(df.shape)
df.head()
````

````python
df.info()
````

````python
df.describe()
````

````python
df.isnull().sum()
````

````python
cleaned = df.drop(['name','ticket','fare','cabin','embarked','boat','body','home.dest'], axis=1, errors='ignore')
cleaned['age'] = cleaned['age'].fillna(cleaned.groupby('pclass')['age'].transform('mean'))
cleaned.isnull().sum()
````

````python
sns.catplot(x='sex', hue='survived', kind='count', data=cleaned)
plt.title('Survival Count by Sex')
plt.show()
````

````python
group1 = cleaned.groupby(['sex','survived'])
gender_survived = group1.size().unstack()
sns.heatmap(gender_survived, annot=True, fmt='d')
plt.title('Gender vs Survived')
plt.show()
````

````python
group2 = cleaned.groupby(['pclass','survived'])
pclass_survived = group2.size().unstack()
sns.heatmap(pclass_survived, annot=True, fmt='d')
plt.title('Pclass vs Survived')
plt.show()
````

````python
sns.violinplot(x='sex', y='age', hue='survived', data=cleaned, split=True)
plt.title('Age Distribution by Sex & Survival')
plt.show()
````

````python
corr_df = cleaned.drop(['sex'], axis=1, errors='ignore')
corr_df = corr_df.apply(pd.to_numeric, errors='coerce')
sns.heatmap(corr_df.corr(method='pearson'), annot=True, vmax=1)
plt.title('Pearson Correlation Heatmap')
plt.show()
````
