## 4A - Predict salary from experience

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression

x, y, coef = datasets.make_regression(
    n_samples=100,
    n_features=1,
    n_informative=1,
    noise=10,
    coef=True,
    random_state=0
)

x = np.interp(x, (x.min(), x.max()), (0, 20))
print(len(x))
print(x)

y = np.interp(y, (y.min(), y.max()), (20000, 150000))
print(len(y))
print(y)

plt.plot(x, y, ".", label="training data")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()

reg_model = LinearRegression()
reg_model.fit(x, y)
y_predicted = reg_model.predict(x)

plt.plot(x, y_predicted, ".", color="black", label="predicted data")
plt.plot(x, y, ".", label="training data")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.legend()
plt.show()
```

In 4A, synthetic data is generated and scaled so it looks like years of experience and salary values.  
A linear regression model is trained on this data, and then the predicted points are plotted with the original data.  
This helps visualize how a straight line can model the relationship between experience and salary.

## 4B - Simulate linear model `Y = 10 + 7x + e`

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()
x = np.random.rand(100, 1)
error = np.random.rand(100, 1)
b0 = 10
b1 = 7
y = b0 + b1 * x + error

reg_model.fit(x, y)
y_predicted = reg_model.predict(x)

plt.plot(x, y_predicted, ".", color="black", label="Predicted data")
plt.plot(x, y, ".", label="Training data")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X vs Y")
plt.legend()
plt.show()
```

In 4B, random `x` values are created and the formula `Y = 10 + 7x + e` is used to generate output values.  
The model learns this pattern from the generated data and predicts new values using linear regression.  
The final graph shows that the predicted values closely follow the training data.
